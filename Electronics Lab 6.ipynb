```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = np.array([100,1000,10000,100000,1000000])
V_in = np.array([2.04,2.04,2.04,2.04,2.04])
V_out = np.array([1.46,1.48,1.46,1.44,1.45])
G = V_out/V_in

#p = f.argsort()
#f = f[p]
#G = G[p]

dB = 20*np.log10(G)
```

```python
fig, ax = plt.subplots(1, 1)

fig.set_figheight(10)
fig.set_figwidth(10)
ax.set_xscale("log")
ax.plot(f, dB, "o-", label=r"$20log\frac{V_{out}}{V_{in}}$")
#ax.set_xlim(10**1.9, 10**3.5)
ax.set_title("Decibels vs Frequency")
plt.xlabel("Frequency $f$ (Hz)")
plt.ylabel("Decibels $(dB)$")
#plt.hlines(-3, 10**1, 10**3.5, linestyles="--", color="BLACK", label=r"-3dB")
#plt.vlines(944, -20, 1, linestyles="--", color="RED", label="944 Hz")
ax.set_ylim(-10, 1)
ax.legend(prop={'size': 16})
plt.show()
```

```python
a = np.arange(20, 1000, 10)
b = np.arange(1000, 12200, 100)
c = np.append(a, b)
f = np.append(c, [12500, 13000, 13500, 14000, 14500, 15100, 15200, 15400, 15600, 15800, 16000, 16500, 17000, 17500, 18000, 18400, 19000, 19500, 20000])
d = np.array([.624, .915, 1.21, 1.46, 1.66, 1.87, 1.66, 1.46, 1.41, 1.46, 1.46, 1.46, 1.46, 1.50, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.6, 1.6, 1.6, 1.62, 1.64, 1.64, 1.64], dtype=float)
e = np.full(132, fill_value=1.66, dtype=float)
g = np.array([1.64, 1.64, 1.64, 1.64, 1.64, 1.62, 1.6, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.58, 1.54, 1.54, 1.50, 1.46, 1.46, 1.46, 1.46, 1.46, 1.41, 1.39, 1.37, 1.37, 1.37, 1.37, 1.37, 1.37, 1.37, 1.37, 1.29, 1.29, 1.29, 1.29])
h = np.append(d, e)
V_out = np.append(h, g)
V_in = np.full(f.size, fill_value=0.216, dtype=float)
G = V_out/V_in

#p = f.argsort()
#f = f[p]
#G = G[p]

dB = 20*np.log10(G)
```

```python
fig, ax = plt.subplots(1, 1)

fig.set_figheight(10)
fig.set_figwidth(10)
ax.set_xscale("log")
ax.plot(f, dB, "o-", label=r"$20log\frac{V_{out}}{V_{in}}$")
ax.set_xlim(10**1, 10**5)
ax.set_title("Decibels vs Frequency")
plt.xlabel("Frequency $f$ (Hz)")
plt.ylabel("Decibels $(dB)$")
plt.hlines(14.7131, 10**1, 10**5, linestyles="--", color="BLACK", label=r"-3dB")
#plt.vlines(944, -20, 1, linestyles="--", color="RED", label="944 Hz")
ax.set_ylim(5, 22)
ax.legend(prop={'size': 16})
plt.show()
```

```python
R = 4
Vrms = V_out/(2*np.sqrt(2))
P = (Vrms**2)/R
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = np.array([10, 100, 1000, 10000, 15, 20, 25, 30, 35, 40, 45, 50, 55, 65, 7])
V_in = np.array([0.996, 1.01, 1.02, 1.02, 1.02, 1.02, 1.00, 1.01, 1.02, 1.01, 1.02, 1.02, 1.01, 1.02, 1.02])
V_out = np.array([3.32, 6.16, 6.16, 6.16, 4.44, 5.00, 5.00, 5.68, 5.84, 5.88, 6.00, 6.00, 6.04, 6.16, 2.30])
G = V_out/V_in

p = f.argsort()
f = f[p]
G = G[p]

dB = 20*np.log10(G)
```

```python
fig, ax = plt.subplots(1, 1)

fig.set_figheight(10)
fig.set_figwidth(10)
ax.set_xscale("log")
ax.plot(f, dB, "o-", label=r"$20log\frac{V_{out}}{V_{in}}$")
#ax.set_xlim(10**1.9, 10**3.5)
ax.set_title("Decibels vs Frequency")
plt.xlabel("Frequency $f$ (Hz)")
plt.ylabel("Decibels $(dB)$")
#plt.hlines(-3, 10**1, 10**3.5, linestyles="--", color="BLACK", label=r"-3dB")
#plt.vlines(944, -20, 1, linestyles="--", color="RED", label="944 Hz")
#ax.set_ylim(-16.5, 0.5)
ax.legend(prop={'size': 16})
plt.show()
```
